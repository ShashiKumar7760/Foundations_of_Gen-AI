```python
import os
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate

# ======================================
# 1. Set Groq API Key
# ======================================
os.environ["GROQ_API_KEY"] = "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"

# ======================================
# 2. Load Groq LLM
# ======================================
llm = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0.7
)

# ======================================
# 3. Prompt Template
# ======================================
prompt = PromptTemplate(
    input_variables=["question"],
    template="""
You are a helpful AI assistant.
Answer the question clearly and simply.

Question: {question}
"""
)

# ======================================
# 4. Create Runnable Chain (NEW WAY)
# ======================================
chain = prompt | llm

# ======================================
# 5. User Interactive Q&A
# ======================================
print("\n=== LangChain + Groq User Q&A App ===")

while True:
    question = input("\nAsk a question (type 'exit' to quit): ")

    if question.lower() == "exit":
        print("Session ended.")
        break

    if question.strip() == "":
        print("Please enter a valid question.")
        continue

    response = chain.invoke({"question": question})
    print("\nAnswer:")
    print(response.content)
```
